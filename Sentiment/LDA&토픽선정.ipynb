```python
import gensim
from gensim import corpora, models
from gensim.corpora import Dictionary
```

### 1.1 전체 단어의 사전 만들고 각 문서에 매칭하기

```python
df_cluster1=df[df['cluster']==1]
df_cluster1.head()
```

```python
all_document = list(df_cluster1['tagged_review'])
dictionary = Dictionary(all_document) #단어 사전 생성기기
dictionary.token2id
```

```python
corpus = []
for doc in all_document:
    corpus.append(dictionary.doc2bow(doc))
```

```python
corpus
```

* 2.1 LDA 기본 모델 만들기

### 2.1 LDA 기본 모델 만들기

```python
#passes 미설정시 기본값 1
#itereations 미설정시 기본값 50
lda_model = gensim.models.ldamodel.LdaModel(corpus,
                                           num_topics=3,
                                           id2word=dictionary,
                                           passes=5, #수렴과정을 몇번 거칠지
                                           iterations=200, #한번에 몇개의 문서를 처리할지
                                           random_state=1)
```

```python
lda_model.print_topics(num_words=5)
```

### 2.2 LDA 토픽개수 선정
* cohearence & perplexcity

```python
from gensim.models import CoherenceModel
import matplotlib.pyplot as plt
import numpy as np
```

```python
perplexity_score = lda_model.log_perplexity(corpus)
perplexity_score
```

```python
coherence_model =CoherenceModel(model = lda_model, texts=all_document, dictionary=dictionary)
coherence_model.get_coherence()
```

> 2.2.1 Perplexity 그래프 만들기

```python
p_s_list = []
c_s_list = []

for i in tqdm(range(2, 11)):
    lda_model = gensim.models.ldamodel.LdaModel(corpus,
                                           num_topics=i,
                                           id2word=dictionary,
                                           passes=1, #수렴과정을 몇번 거칠지
                                           iterations=25, #한번에 몇개의 문서를 처리할지
                                           random_state=1)
    
    p_s = lda_model.log_perplexity(corpus) #perplexity score
    p_s_list.append(p_s)
    
    #coherence_score
    coherence_model =CoherenceModel(model = lda_model, texts=all_document, dictionary=dictionary)
    c_s = coherence_model.get_coherence()
    c_s_list.append(c_s)
```

```python
#perplexity score 그래프 만들기
x = range(2,11)
plt.plot(x, p_s_list)
```

> 2.2.1 Cohearence 그래프 만들기

```python
x = range(2,11)
plt.plot(x, c_s_list)
```

### 2.3 선정한 토픽 개수로 Action 넘버 매칭

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus,
                                       num_topics=4,
                                       id2word=dictionary,
                                       passes=5, #수렴과정을 몇번 진행할지
                                       iterations=200, #한번에 몇개의 문서를 처리할지
                                       random_state=1) #random state를 설정하면 항상 같은 예측값이 나온다.
```

```python
#각 document들에 들어가 있는 토픽들이 배당될 확률 출력
lda_model.get_document_topics(corpus)[566]
```

```python
#label - document에 대한 토픽 넘버(=action number)
#value - 각 토픽에 배당될 확률

action_cluster_list =[] #가장 높은 확률값을 가진 토픽 넘버들의 모음
for i in lda_model.get_document_topics(corpus):
    label = [] #0 1 2 3 
    value = []#0.5 0.2 0.02 0.41
    for t, s in i:
        label.append(t)
        value.append(s)

    index = np.argmax(value) #value 리스트(확률값들만 모여있는 리스트)에서 가장 값이 큰 요소의 인덱스 반
    action_cluster =label[index] #label 데이터(토픽넘버들만 모아둔 리스트)에서 반환된 인덱스로 인덱싱
    action_cluster_list.append(action_cluster) 
```

```python
action_cluster_list
```

```python
df_cluster1['action_cluster'] = action_cluster_list
df_cluster1.head()
```
